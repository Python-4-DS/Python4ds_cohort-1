```python
loan_reg = pd.merge(loan, mpi_region_location).dropna()
loan_reg
```

```python
#Cleaning the gender borrower_gender

#loan_reg['gender'] = loan_reg['borrower_genders'].apply(lambda x: x.split(', ')[0])
#loan_reg['gender'] = loan_reg['borrower_genders'].apply(lambda x: ', '.join(set(x.lower().split(', '))))
#loan_reg.drop(['female ,male'])
#Loan_filtered = loan_reg[loan_reg['gender'] != 'female, male']
#Loan_filtered.T
loan_reg['gender'] = loan_reg['borrower_genders'].apply(lambda x: x.split(',')[0])
loan_reg.head(100).T
```

```python
#percentage by gender; Overall
loan_reg['gender'].value_counts(normalize=True) * 100
```

```python
#percentage by gender; by country
loan_reg.groupby('country')['gender'].value_counts(normalize=True) * 100
```

```python
#percentage by gender; by world  region
loan_reg.groupby('world_region')['gender'].value_counts(normalize=True) * 100
```

```python
#percentage by gender; by region
loan_reg.groupby('region')['gender'].value_counts(normalize=True) * 100
```

```python
#Gender with the highest lend count
loan_reg.groupby('gender')['lender_count'].count()
```

```python
#Regions with the highest and lowest loan_amount, funded amount
loan.groupby('region')['loan_amount', 'funded_amount'].sum().sort_values(by='loan_amount',ascending=False)
```

```python
#countries with the highest and lowest loan_amount, funded amount
loan.groupby('country')['loan_amount', 'funded_amount'].sum().sort_values(by='loan_amount', ascending=False)
```

```python
#high lender count
loan['country'].value_counts().reset_index().head(10)
```

```python
#finding the mean on lender count
loan.groupby(['country']).agg({'lender_count':'mean'}).sort_values(by='lender_count',ascending=False)
```
