# 7. Data Preprocessing
<a id="preprocessing"></a>
Here we extract the features and the target class.

```python
# make a copy of the datasets
training = santander_training.copy()
test = santander_test.copy()
```

```python
# Drop unwanted columns in training set
training_set = training.drop(columns=['ID_code'], axis=1)
training_set.head()
```

```python
# Drop unwanted columns in test set
test_set = test.drop(columns=['ID_code'], axis=1)
test_set.head()
```

```python
# Extract the target class and drop it. Extract x.
y = training_set['target'].values
training_set.drop(columns=['target'], axis=1, inplace=True)
training_set.head()
```

```python
X = training_set.values
# print(X)
```

# 8. Feature Engineering
<a id="f-eng"></a>
Feature engineering is important for better model performance.
Looking at our data, the features are not uniform scalewise. This calls for feature scaling to make sure they are uniform

## 8.1. Scaling the Data
<a id="scaling"></a>
We'll use scikit learn's StandardScaler class.
Before scaling, we need to split the training set into Folds

```python
# Instantiate the scaler
scaler = StandardScaler()

# Transform training data
X = scaler.fit_transform(X)

# Transform test_set
X_test = test_set.values
X_test = scaler.transform(X_test)
```

# 9. Model Selection & Training
<a id="selection"></a>
Here we'll train several models with the training data and choose the best performing model.

## 9.1. Choosing a performace metric
<a id="metric"></a>
For choosing the best model, we'll use K-fold cross validation of 10 folds and compute the mean score of each fold.

## 9.2. Model Evaluation
<a id="evaluation"></a>
We'll compare the score of 4 models

```python
# Create a score collection for comparison
scores = {}
```

### Gradient Boosting Classifier
We implement a Gradient Boosting classifier with a max_depth of 1 and 0 random state

```python
GBC = GradientBoostingClassifier(max_depth=1, random_state=0)

GBC_score = cross_val_score(GBC, X, y, cv=10, scoring='accuracy').mean()

print(GBC_score)

scores['GBC'] = GBC_score
```

### Random Forest Classifier
We'll implement another ensemble, Random Forest Classifier, with default parameters(10 estimators)

```python
RFC = RandomForestClassifier(n_estimators=10)

RFC_score = cross_val_score(RFC, X, y, cv=10, scoring='accuracy').mean()

print(RFC_score)

scores['RFC'] = RFC_score
```

### MLP Classifier
Here we'll implement a neural network, MLP Classifier

```python
MLP = MLPClassifier(hidden_layer_sizes=(10, 10), random_state=0, max_iter=1000)

MLP_score = cross_val_score(MLP, X, y, cv=10, scoring='accuracy').mean()

print(MLP_score)

scores['MLP'] = MLP_score
```

### Kernelized SVM
Here we'll implement a kernelized Support vector machine with a Gaussian model kernel

```python
svc = SVC(kernel='rbf', random_state=0, gamma='auto')

svc_score = cross_val_score(svc, X, y, cv=10, scoring='accuracy').mean()

print(svc_score)

scores['svc'] = svc_score
```

#### Lets look at the score for each model

```python
scores
```

Its evident that support vector classifier performed well. We'll fine tune its parameters for optimization

# 10. Model Evaluation and Optimization
<a id="evaluation"></a>
We'll use GridSearchCV to fine tune the model.

```python
# Initialize the tuning params
Cs = [0.001, 0.01, 0.1, 1, 10]
gammas = [0.001, 0.01, 0.1, 1, 10]

param_grid = {'C': Cs, 'gamma': gammas}

gridscv = GridSearchCV(
    estimator=SVC(kernel='rbf', random_state=0),
    param_grid=param_grid,
    scoring='accuracy',
    n_jobs=1,
    cv=10
)
```

# 11. Prediction
<a id=""></a>
Here we fit the model with the fine tuned params and make predictions

```python
model = SVC(kernel='rbf', random_state=0, gamma='auto')

model.fit(X, y)
```

```python
predictions = model.predict(X_test)

prediction_df = pd.DataFrame(predictions)

prediction_df.head(50)
```
